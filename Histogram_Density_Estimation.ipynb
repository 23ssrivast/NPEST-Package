```python
import numpy as np
import matplotlib.pyplot as plt
import npest
np.random.seed(42)
data = np.concatenate([
    np.random.normal(-2, 1, 300),
    np.random.normal(3, 1.5, 200)
])
methods = ['sturges', 'fd', 'scott', 'blocks']

plt.figure(figsize=(12, 8))
for i, method in enumerate(methods, 1):
    hist = npest.Histogram(bins=method, density=True).fit(data)
    n_bins = len(hist.hist_)
    x = np.linspace(data.min(), data.max(), 500)
    y = hist.pdf(x)

    plt.subplot(2, 2, i)
    plt.hist(data, bins=40, density=True, alpha=0.5, color='gray')
    plt.plot(x, y, 'b-', lw=2, label=f'{method} ({n_bins} bins)')
    plt.title(f'Histogram – {method}')
    plt.legend()

plt.tight_layout()
plt.show()
```
