```python
import numpy as np
import npest

np.random.seed(42)
sample = np.random.normal(0, 1.5, 100)

p_two_sided = npest.bootstrap_test(np.var, sample, null_value=1.0,
                                   n_bootstrap=1000, alternative='two-sided')
p_greater = npest.bootstrap_test(np.var, sample, null_value=1.0,
                                 n_bootstrap=1000, alternative='greater')

print("Sample variance = {:.3f}".format(np.var(sample)))
print("\nTesting H0: variance = 1.0")
print("  Two‑sided p‑value : {:.4f}".format(p_two_sided))
print("  Greater p‑value   : {:.4f}".format(p_greater))

sample_good = np.random.normal(0, 1, 100)
p_good = npest.bootstrap_test(np.var, sample_good, null_value=1.0,
                              n_bootstrap=1000, alternative='two-sided')

print("\nFor a sample with true variance = 1.0:")
print("  Sample variance = {:.3f}".format(np.var(sample_good)))
print("  Two‑sided p‑value = {:.4f}".format(p_good))
```
