```python
import numpy as np
import npest

np.random.seed(42)
x = np.random.normal(0, 1, 200)
y = np.random.normal(0.8, 1.2, 200)
z = np.random.normal(0, 1, 200)
stat_ks, p_ks = npest.two_sample_test(x, y, method='ks', n_permutations=500)
stat_kde, p_kde = npest.two_sample_test(x, y, method='kde', n_permutations=500)

print("Testing x vs. y (different means):")
print(f"  KS test  : statistic = {stat_ks:.3f}, p‑value = {p_ks:.4f}")
print(f"  KDE test : statistic = {stat_kde:.3f}, p‑value = {p_kde:.4f}")

stat_ks_same, p_ks_same = npest.two_sample_test(x, z, method='ks', n_permutations=500)
stat_kde_same, p_kde_same = npest.two_sample_test(x, z, method='kde', n_permutations=500)

print("\nTesting x vs. z (same distribution):")
print(f"  KS test  : statistic = {stat_ks_same:.3f}, p‑value = {p_ks_same:.4f}")
print(f"  KDE test : statistic = {stat_kde_same:.3f}, p‑value = {p_kde_same:.4f}")
```
