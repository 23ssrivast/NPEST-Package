```python
import numpy as np
import matplotlib.pyplot as plt
import npest

np.random.seed(42)
data1 = np.random.normal(-2, 1, 300)
data2 = np.random.normal(3, 1.5, 200)
data = np.concatenate([data1, data2])

kernels = ['gaussian', 'epanechnikov', 'triangular', 'box']

plt.figure(figsize=(12, 8))
for i, kernel in enumerate(kernels, 1):
    kde = npest.KDE(bandwidth='silverman', kernel=kernel).fit(data)
    x = np.linspace(data.min()-1, data.max()+1, 500)
    y = kde.pdf(x)

    plt.subplot(2, 2, i)
    plt.hist(data, bins=40, density=True, alpha=0.5, color='gray', label='Histogram')
    plt.plot(x, y, 'r-', lw=2, label=f'{kernel} KDE')
    plt.title(f'KDE – {kernel} kernel (bw = {kde.bw_:.3f})')
    plt.legend()

plt.tight_layout()
plt.show()
```
