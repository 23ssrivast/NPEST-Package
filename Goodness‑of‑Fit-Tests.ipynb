```python
import numpy as np
import npest

np.random.seed(42)
data_t = np.random.standard_t(df=5, size=200)
data_normal = np.random.normal(0, 1, 200)
p_ks_t = npest.goodness_of_fit(data_t, dist='normal', method='ks')
p_kde_t = npest.goodness_of_fit(data_t, dist='normal', method='kde', n_bootstrap=500)

print("Testing t‑distributed data vs. normal:")
print(f"  KS test p‑value  : {p_ks_t:.4f}")
print(f"  KDE test p‑value : {p_kde_t:.4f}")

p_ks_norm = npest.goodness_of_fit(data_normal, dist='normal', method='ks')
p_kde_norm = npest.goodness_of_fit(data_normal, dist='normal', method='kde', n_bootstrap=500)

print("\nTesting normal data vs. normal:")
print(f"  KS test p‑value  : {p_ks_norm:.4f}")
print(f"  KDE test p‑value : {p_kde_norm:.4f}")
```
