```python
import numpy as np
import matplotlib.pyplot as plt
import npest
np.random.seed(42)
x = np.random.uniform(0, 10, 100)
y = 2 * x + np.random.normal(0, 3, 100)  
spearman_corr, spearman_p = npest.spearman(x, y)
kendall_corr, kendall_p = npest.kendall(x, y)

print(f"Spearman rank correlation : ρ = {spearman_corr:.3f}, p = {spearman_p:.4f}")
print(f"Kendall's tau             : τ = {kendall_corr:.3f}, p = {kendall_p:.4f}")
try:
    hoeff = npest.hoeffding(x, y)
    print(f"Hoeffding's D             : {hoeff.statistic:.3f}, p = {hoeff.pvalue:.4f}")
except NotImplementedError as e:
    print(f"Hoeffding's D             : {e}")
plt.figure(figsize=(6, 5))
plt.scatter(x, y, alpha=0.7, edgecolors='k')
plt.title(f'Correlated data\nSpearman ρ = {spearman_corr:.3f}')
plt.xlabel('x')
plt.ylabel('y')
plt.tight_layout()
plt.savefig('correlation_example.png')
plt.show()
```
